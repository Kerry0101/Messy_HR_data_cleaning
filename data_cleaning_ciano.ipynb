```python
# Fill NaN values in the Age column with a constant value
# You can replace 30 with any other value you prefer
df['Age'] = df['Age'].astype(str).replace('30', '')

# Display the first few rows of the dataframe to verify
df.head()
```

```python
# Function to convert spelled-out numbers to numerical values
def convert_salary_to_numeric(salary):
    if pd.isna(salary):
        return salary
    
    # Convert to string in case it's already numeric
    salary_str = str(salary).upper()
    
    # Define mapping for common spelled-out numbers
    number_mapping = {
        'SIXTY THOUSAND': 60000,
        'SEVENTY THOUSAND': 70000,
        'FIFTY THOUSAND': 50000,
        'FORTY THOUSAND': 40000,
        'EIGHTY THOUSAND': 80000,
        'NINETY THOUSAND': 90000
        # Add more mappings as needed
    }
    
    # Check if the salary is in our mapping
    if salary_str in number_mapping:
        return number_mapping[salary_str]
    
    # If it's not in our mapping but is already numeric or NaN, return as is
    try:
        return float(salary_str)
    except:
        return salary

# Apply the function to the Salary column
df['Salary'] = df['Salary'].apply(convert_salary_to_numeric)
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Salary'] = df['Salary'].fillna(df['Salary']) 
df['Salary'] = df['Salary'].astype('Int64')

```

```python
# Function to convert spelled-out numbers to numerical values
def convert_age_to_numeric(age):
    if pd.isna(age):
        return age
    
    # Convert to string in case it's already numeric
    age_str = str(age).lower()  # Use lower() instead of upper() for "thirty"
    
    # Define mapping for common spelled-out numbers
    number_mapping = {
        'thirty': 30,
        'nan': None,  # Convert "nan" string to actual None/NaN
    }
    
    # Check if the age is in our mapping
    if age_str in number_mapping:
        return number_mapping[age_str]
    
    # If it's not in our mapping but is already numeric or NaN, return as is
    try:
        return float(age_str)
    except:
        return age

# Apply the function to the Age column
df['Age'] = df['Age'].apply(convert_age_to_numeric)

# Convert to numeric and handle any remaining string "nan" values
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Convert to integer format (no decimals)
df['Age'] = df['Age'].astype('Int64')  # Int64 preserves NaN values

# When saving to CSV, represent NaN values as empty strings
df.to_csv('messy_HR_data_cleaned.csv', index=False, na_rep='')
```

```python
# Import the necessary library for date handling
from datetime import datetime
import pandas as pd

# Function to convert text dates to YYYY/MM/DD format
def convert_date_format(date_str):
    if pd.isna(date_str):
        return date_str
    
    try:
        # Try parsing with pandas (handles many formats automatically)
        date_obj = pd.to_datetime(date_str)
        # Format as YYYY/MM/DD
        return date_obj.strftime('%Y/%m/%d')
    except:
        # If pandas can't parse it, return the original
        return date_str

# Apply the function to the 'Joining date' column
df['Joining Date'] = df['Joining Date'].apply(convert_date_format)


# Save the updated DataFrame to CSV
df.to_csv('messy_HR_data_cleaned.csv', index=False, na_rep='')
```
